```python
def shell_sort(lst):
  gap = len(lst) // 2
  while gap > 0:
    for start in range(gap):
      gap_ins(lst, start, gap)
    gap = gap // 2

def gap_ins(lst, start, gap):
  for i in range(start + gap, len(lst), gap):
    val = lst[i]
    pos = i
    while pos >= gap and lst[pos - gap] > val:
      lst[pos] = lst[pos - gap]
      pos = pos - gap
    lst[pos] = val
```

```python
test_list = [64, 34, 25, 12, 22, 11, 90, 5, 78, 1]
print("Before:", test_list)
shell_sort(test_list)
print("After: ", test_list)
```

```python
def merge_sort(lst):
  if len(lst) > 1:
    mid = len(lst) // 2
    left = lst[:mid]
    right = lst[mid:]
    merge_sort(left)
    merge_sort(right)
    i = j = k = 0
    while i < len(left) and j < len(right):
      if left[i] <= right[j]:
        lst[k] = left[i]
        i += 1
```

```python
test_list = [38, 27, 43, 3, 9, 82, 10, 55, 21, 6]
print("Before:", test_list)
sorted_list = merge_sort(test_list)
print("After: ", sorted_list)
```
