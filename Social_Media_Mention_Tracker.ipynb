```python
# steam of posts
posts = [
    "I love learning Python with @code_master!",
    "Is @alice coming to the hackathon?",
    "Shoutout to @code_master for the help today.",
    "Hey @bob, did you see the post by @alice?",
    "Python dictionaries are awesome. @code_master agrees!"
]

# create an empty dictionary to store the counts
mentions = {}

# process the ports
for post in posts:
  words = post.split()

  for word in words:
    clean_word = word.strip("!?.,")

    if clean_word.startswith("@"):
      if clean_word in mentions:
        mentions[clean_word] += 1
      else:
        mentions[clean_word] = 1

# display the results
print("---Mention Count ---")
for user, count in mentions.items():
  print(f"{user}: {count} mentions")

# Reflection Question: Why use a dictionary?
# Using a dictionary is very efficient since it stores the data
# by giving it its own, unique key. By having their own keys, this
# makes it so python is able to quickly find and change a username
# without much searching. If you used a list instead, you would
# have to literally check every single item everytime you wanted to
# update a count, which would take a really long time, especially
# if you had 1,000,000 posts. Dictionarys make lookups and updates
# very easy, which are why they are more efficient than lists.
```
